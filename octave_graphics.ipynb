# Graphics in MATLAB/Octave

## Two-dimensional plots
The MATLAB/Octave command `plot(x,y)` creates a 2-D line plot of the data in `x` versus the corresponding values in `y` (for more details, see https://www.mathworks.com/help/matlab/line-plots.html).

### Create a line plot
As an example, first create a vector `x` of linearly spaced values between 0 and $2\pi$, using an increment of $\pi/100$. Then create a vector `y` that contains the sine values of `x`. Finally, create a line plot of the data in `y` versus the corresponding values in `x`. 

```octave
% register_graphics_toolkit ('fltk')
% available_graphics_toolkits()
% graphics_toolkit('fltk')
% %plot --format svg
x = 0:pi/100:2*pi;
y = sin(x);
plot(x,y)
```

### Add title and axis labels to chart
As a variant of the previous example, label the axes and add a title:

```octave
x = 0:pi/100:2*pi;
y = sin(x);
plot(x,y)
xlabel('x')
ylabel('sin(x)')
title('Plot of the Sine Function')
```

```octave
x = linspace(-2*pi,2*pi,100);
y1 = sin(x);
y2 = cos(x);
plot(x,y1,x,y2)
```

### Plot multiple lines
In the next example, create `x` as 100 linearly spaced values between $−2π$ and $2π$. Create `y1` and `y2` as sine and cosine values of x. Plot both sets of data, label the axes and add a title:

```octave
x = linspace(-2*pi,2*pi,100);
y1 = sin(x);
y2 = cos(x);
plot(x,y1,x,y2)
xlabel('-2\pi < x < 2\pi') 
ylabel('Sine and Cosine Values')
title('Line Plot of Sine and Cosine Between -2\pi and 2\pi')
```

To display the Greek symbol $\pi$ in the label of the `x` axis or title, use the TeX markup, `\pi`.

### Add a legend
Now add a legend to the graph to identify each data set in the order that you plot the lines. Optionally, specify the legend location using one of the eight cardinal or intercardinal directions, for example, `'southwest'`:

```octave
x = linspace(-2*pi,2*pi,100);
y1 = sin(x);
y2 = cos(x);
plot(x,y1,x,y2)
xlabel('-2\pi < x < 2\pi') 
ylabel('Sine and Cosine Values')
title('Line Plot of Sine and Cosine Between -2\pi and 2\pi')
legend({'y = sin(x)','y = cos(x)'},'Location','southwest')
```

### Change font size
One way to change the font size is to set the `FontSize` property of the title, axis labels or legend, for example, to increase the font size of the title to 16 pt, modify the `title` command as follows`:

```octave
x = linspace(-2*pi,2*pi,100);
y1 = sin(x);
y2 = cos(x);
plot(x,y1,x,y2)
xlabel('-2\pi < x < 2\pi') 
ylabel('Sine and Cosine Values')
title('Line Plot of Sine and Cosine Between -2\pi and 2\pi', ...
      'FontSize', 16)
legend({'y = sin(x)','y = cos(x)'},'Location','southwest')
```

Note that the ellipsis `...` allows to continue a statement to the next line.

## Three-dimensional plots
The following example shows how to create a three-dimensional plot by creating a mesh. 

First create two column vectors x and y:

```octave
x = linspace (-8, 8, 41)';
y = x;
```

Then use the `meshgrid` function to replicate the grid vectors x and y an produce the coordinates of a rectangular grid (xx, yy).

```octave
[xx, yy] = meshgrid (x, y);
```

Now calculate the function ${\rm sinc}\,x = \frac{\sin\,x}{x}$ as a function of the distance from the center (add the constant `eps` to avoid dividing by zero):

```octave
r = sqrt (xx .^ 2 + yy .^ 2) + eps;
z = sin (r) ./ r;
```

Finally plot ${\rm sinc}\,x$ as a three-dimensional colored mesh surface using the `mesh` function, with the color being proportional to the mesh height `z`:

```octave
mesh (x, y, z);
```
